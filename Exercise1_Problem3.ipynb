```python
#(a)

import numpy as np
import matplotlib.pyplot as plt

x = [1/3, 1/12]
for i in range(1, 59):
    x.append(9/4 * x[i] - 1/2 * x[i-1])

k_values = range(1, 61)
plt.semilogy(k_values, x)
plt.xlabel('k')
plt.ylabel('$x_k$')
plt.title('Semilog plot of the sequence')
plt.show()
```

```python
# (b)

import numpy as np
import matplotlib.pyplot as plt

x = []

for i in range(60):
    x.append(1/3 * 4 ** (1-i))

k_values = range(1, 61)
plt.semilogy(k_values, x)
plt.xlabel('k')
plt.ylabel('$x_k$')
plt.title('Semilog plot of the sequence')
plt.show()
```
