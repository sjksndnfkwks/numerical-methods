```python
import numpy as np
import matplotlib.pyplot as plt

# Define the polynomial function using explicit representation
def p_explicit(x):
    return x**8 - 24*x**7 + 252*x**6 - 1512*x**5 + 5670*x**4 - 13608*x**3 + 20412*x**2 - 17496*x + 6561
    
# Define the polynomial function using compact formula
def p_compact(x):
    return (x - 3)**8

x_values = np.arange(2.920, 3.081, 0.001)

y_explicit = p_explicit(x_values)
y_compact = p_compact(x_values)

# Plotting for part a)
plt.plot(x_values, y_explicit, label='Explicit Representation')
plt.title('Function p(x) using Explicit Representation')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()

# Plotting for part b)
plt.plot(x_values, y_compact, label='Compact Formula')
plt.title('Function p(x) using Compact Formula')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```
